### The Print Function

Functions are interfaces to interact with the compiler to pass input and execute instructions.

The Print function accepts input and executes instructions to show the output.

Syntax:

`print( object(s) )`

The function runs by calling its names and provide the input in the parentheses.

`objects(s)` is the function input.

The comma symbol `,` can be used to separate objects.

### Print Examples

```python
print(5)

# Printing two objects
print(5, 7)

# Printing a result of an arithmetic operation
print(5 + 2)

print(7 * 3)
```

### Arithmetic Operations in Python

- Python realizes the order of operations
- Multiplication and division have higher priority than addition and subtraction
- Parentheses have the highest priority
- In addition to subtraction `-`, addition `+`, division `\`, and multiplication `*`, you can perform the following:
- Double asterisk `**` As exponentiation operator
- Double forward slashes `//` to find quotient without remainder



### Examples

```python
# Parentheses used
print( (6 + 2) * 4 / 2 )

# Exponential operator
print( 3 ** 2 )

# Exponential operator again
print( 9 ** 0.5 )

# Quotient without remainder
print( 13 // 3 )
```
